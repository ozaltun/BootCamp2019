```python
def get_bernstein_basis(n,v, x):
    return special.binom(n, v) * (x**v) * (1-x)**(n-v)
```

```python
x = np.linspace(0, 1, 1000)
```

```python
fig, axes = plt.subplots(4, 4, sharey=True, figsize=(10, 5))

for n in range(4):
    v = 0
    while v <= n:
        axes[n, v].plot(x, get_bernstein_basis(n,v,x), label="n = "+str(n))
        axes[n, v].set_title('n={}, v={}'.format(n,v))
        v += 1

plt.subplots_adjust(hspace=1.5)
fig.suptitle('Bernstein basis functions')
plt.show()
```

```python
def rosenbrock_func(x, y):
    return (1-x)**2 + 100*(y-x**2)**2
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

x = np.linspace(-8,10)
y = np.linspace(-8,10)

X, Y = np.meshgrid(x,y)
Z = np.zeros((50,50))

for i in range(50):
    for j in range(50):
        Z[i,j] = rosenbrock_func(X[i,j],Y[i,j])
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='viridis',edgecolor='none')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='magma',edgecolor='none')
```

```python
plt.contourf(X, Y, Z, cmap='viridis')
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='viridis',edgecolor='none')

Z2 = np.zeros((50,50))
ax.plot_surface(X, Y, Z2, rstride=1, cstride=1,cmap='magma',edgecolor='none')
ax.scatter(1, 1, 0, s=100)
```

```python
data2 = np.load('countries.npy')
```

```python
countries = ["Austria", "Bolivia", "Brazil", "China",
            "Finland", "Germany", "Hungary", "India",
            "Japan", "North Korea", "Montenegro", "Norway",
            "Peru", "South Korea", "Sri Lanka", "Switzerland",
            "Turkey", "United Kingdom", "United States", "Vietnam"]
```

```python
fig, ax = plt.subplots()
ax.scatter(data2[:,0], data2[:,1])
ax.set_xlabel('Population (in millions)')
ax.set_ylabel('GDP in billions of USD')
for i, txt in enumerate(countries):
    ax.annotate(txt, (data2[i,0], data2[i,1]))
```

```python
fig, ax = plt.subplots()
ax.scatter(data2[:,2], data2[:,3])
ax.set_xlabel('Average male height')
ax.set_ylabel('Average female height')
for i, txt in enumerate(countries):
    ax.annotate(txt, (data2[i,2]+0.5, data2[i,3]))
```

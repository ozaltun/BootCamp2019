```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Problem 1

```python
DJIA = pd.read_csv('DJIA.csv')
```

```python
DJIA['DATE'] = pd.to_datetime(DJIA['DATE'])
DJIA['VALUE'] = pd.to_numeric(DJIA['VALUE'], errors='coerce')
DJIA = DJIA.dropna()
DJIA = DJIA.reset_index(drop=True)
```

```python
DJIA.plot(x='DATE', y = 'VALUE', lw=0.5)
plt.show()
```

## Problem 2

```python
paychecks = pd.read_csv('paychecks.csv', names=['Value'], header=None)
```

```python
paychecks['Date'] = pd.date_range(start='3/13/2008', periods=47, \
        freq='WOM-3FRI').union(pd.date_range(start='3/13/2008', \
        periods=46, freq='WOM-1FRI'))

paychecks['Value']=pd.to_numeric(paychecks['Value'])
```

```python
paychecks.plot(x='Date', y = 'Value')
plt.show()
```

## Problem 5

```python
## TODO
DJIA = DJIA.set_index('DATE')
```

```python
DJIA_daily = DJIA- DJIA.shift(1, freq="D")
```

```python
print('Largest daily increase on ', DJIA_daily['VALUE'].idxmax(), \
                              'with value of ',DJIA_daily['VALUE'].max())
print('Largest daily decrease on ', DJIA_daily['VALUE'].idxmin(),  \
                              'with value of ',DJIA_daily['VALUE'].min())
DJIA_daily.plot()
```

```python
DJIA_monthly = DJIA.groupby(pd.Grouper(freq='1M')).last() - \
                DJIA.groupby(pd.Grouper(freq='1M')).first()

print('Largest monthly increase on ', DJIA_monthly['VALUE'].idxmax(), \
                      'with value of ',DJIA_monthly['VALUE'].max())
print('Largest monthly decrease on ', DJIA_monthly['VALUE'].idxmin(), \
                      'with value of ',DJIA_monthly['VALUE'].min())

DJIA_monthly.plot()
```

## Problem 6

```python
w = 30
ax1 = plt.subplot(221)
DJIA.plot(color="gray", lw=.3, ax=ax1)
DJIA.rolling(window=w).mean().plot(color='r', lw=1, ax=ax1)
ax1.legend(["Actual", "Rolling"], loc="lower right")
ax1.set_title("Rolling Average")

ax2 = plt.subplot(222)
DJIA.plot(color="gray", lw=.3, ax=ax2)
DJIA.ewm(span=w).mean().plot(color='g', lw=1, ax=ax2)
ax2.legend(["Actual", "EWMA"], loc="lower right")
ax2.set_title("EWMA")
plt.tight_layout()

ax2 = plt.subplot(223)
DJIA.plot(color="gray", lw=.3, ax=ax2)
DJIA.rolling(window=w).min().plot(color='g', lw=1, ax=ax2)
ax2.legend(["Actual", "Min Rolling"], loc="lower right")
ax2.set_title("Min Rolling")
plt.tight_layout()

ax2 = plt.subplot(224)
DJIA.plot(color="gray", lw=.3, ax=ax2)
DJIA.rolling(window=w).max().plot(color='g', lw=1, ax=ax2)
ax2.legend(["Actual", "Max Rolling"], loc="lower right")
ax2.set_title("Max Rolling")
plt.tight_layout()
```

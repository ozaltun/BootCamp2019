## Problem 2

```python
def get_bernstein_basis(n,v, x):
    return special.binom(n, v) * (x**v) * (1-x)**(n-v)
```

```python
x = np.linspace(0, 1, 1000)
```

```python
fig, axes = plt.subplots(4, 4, sharey=True, figsize=(10, 5))

for n in range(4):
    v = 0
    while v <= n:
        axes[n, v].plot(x, get_bernstein_basis(n,v,x), label="n = "+str(n))
        axes[n, v].set_title('n={}, v={}'.format(n,v))
        v += 1

plt.subplots_adjust(hspace=1.5)
fig.suptitle('Bernstein basis functions')
plt.show()
```

The Rosenbrock function is defined as follows. 

$$f(x,y)=(1âˆ’x)^2 +100(yâˆ’x^2)^2$$

The minimum value of f is 0, which occurs at the point (1, 1) at the bottom of a steep, banana- shaped valley of the function. Use a heat map and a contour plot to visualize the Rosenbrock function. Plot the minimizer (1, 1). Use a different sequential colormap for each visualization.

```python
def rosenbrock_func(x, y):
    return (1-x)**2 + 100*(y-x**2)**2
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d

x = np.linspace(-8,10)
y = np.linspace(-8,10)

X, Y = np.meshgrid(x,y)
Z = np.zeros((50,50))

for i in range(50):
    for j in range(50):
        Z[i,j] = rosenbrock_func(X[i,j],Y[i,j])
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap='viridis',edgecolor='none')
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
from pydataset import data
import matplotlib.pyplot as plt
```

## Problem 1

```python
titles = ['Titanic', 'birthdeathrates']
datasets = {i:data(i) for i in titles}
```

```python
[print(i, '\n', datasets[i].info()) for i in titles]
```

### Titanic

```python
titanic = datasets['Titanic'][datasets['Titanic'].Freq != 0]
```

```python
titanic.groupby('Sex').sum().plot(kind='barh')
titanic.groupby('Class').sum().plot(kind='barh')
titanic.groupby('Age').sum().plot(kind='barh')
titanic.groupby('Survived').sum().plot(kind='barh')
titanic.groupby('Age').sum().plot(kind='barh')
```

These visualizations help understand the composition of the titanic dataset.

```python
titanic.groupby(['Sex', 'Survived']).sum().plot(kind='barh', figsize=(10,10))
```

```python
titanic.groupby(['Age', 'Survived']).sum().plot(kind='barh', figsize=(10,10))
```

```python
titanic.groupby(['Class', 'Survived']).sum().plot(kind='barh', figsize=(10,10))
```

These last three graphs give a good understanding if there was any priority in saving any group of passengers. It can be seen that 1st class passengers were more likely to be saved.

## birthdeathrates

```python
birthdeathrates = datasets['birthdeathrates'].reset_index()
birthdeathrates.columns = ['Country', 'Birth rate', 'Death rate']
birthdeathrates
```

```python
birthdeathrates.plot(kind='scatter',x='Birth rate', y='Death rate'\
                                , figsize=(15, 15))
```

```python
birthdeathrates.sort_values(by='Birth rate').plot(kind='bar',x='Country', \
                            y='Birth rate' , figsize=(20, 15))
```

```python
birthdeathrates.sort_values(by='Death rate').plot(kind='bar',x='Country', \
                            y='Death rate' , figsize=(20, 15))
```

The first visualization shows the clustering in birth/death rates between countries, the second and third bar plots show the distribution of birth/death rates for each country in increasing order.

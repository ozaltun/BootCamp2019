## Problem 1

```python
def p_1(n):
    mat_0 = np.random.normal(size=(n, n))
    mean_array = np.mean(mat_0, axis = 1)
    var = np.var(mean_array)
    return var

def p_1_part_2():
    list0 = []
    for i in range(1, 11):
        n = i*100
        list0.append(p_1(n))
    return list0
x = [100*i for i in list(range(1,11))]
plt.plot(x, p_1_part_2())
plt.xlabel('n')
plt.ylabel('Variance')
plt.title('Variance for different sizes of matrices ')
plt.show()
```

```python
import matplotlib.gridspec as gridspec

x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Creates two subplots and unpacks the output array immediately
fig, axes = plt.subplots(2, 2, sharey=True)
axes[0, 0].plot(x, np.sin(x), color='green')
axes[0, 0].set_title('sin(x)')
axes[0, 0].set_xlim([0, np.pi*2])
axes[0, 0].set_ylim([-2, 2])

axes[0, 1].plot(x, np.sin(2*x),color='red',linestyle='dashed')
axes[0, 1].set_title('sin(2x)')
axes[0, 1].set_xlim([0, np.pi*2])
axes[0, 1].set_ylim([-2, 2])


axes[1, 0].plot(x,2* np.sin(x),color='blue',linestyle='dashed')
axes[1, 0].set_title('2sin(x)')
axes[1, 0].set_xlim([0, np.pi*2])
axes[1, 0].set_ylim([-2, 2])


axes[1, 1].plot(x, 2*np.sin(2*x),color='magenta')
axes[1, 1].set_title('2sin(2x)')
axes[1, 1].set_xlim([0, np.pi*2])
axes[1, 1].set_ylim([-2, 2])

plt.subplots_adjust(hspace=.5)
fig.suptitle('Plots of various sin functions')
plt.show()
```
